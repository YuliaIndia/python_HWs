###Задача 40: 
Работать с файлом california_housing_train.csv, который находится в папке sample_data. 
Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)

```python
import pandas as pd
df=pd.read_csv('sample_data/california_housing_train.csv')
```

```python
df.describe()
```

```python
df[df['population']<500][['median_house_value']]
```

```python
print(df[df['population']<500][['median_house_value']].mean())
```

```python
print(df['median_house_value'].mean())
```
